WEBSCRAPING PYTHON JOBS FROM A WEBSITE

```python
from bs4 import BeautifulSoup
import requests
%pip install lxml
```

```python
html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=as&searchTextText=Python&txtKeywords=Python&txtLocation=').text
```

```python
html_text
```

```python
soup = BeautifulSoup(html_text, 'lxml')
```

```python
# finding the first match of jobs in the page
jobs = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx')
#print(job)
```

```python
company_name = job.find('h3', class_ = 'joblist-comp-name').text.strip()
print(company_name)
```

```python
skills = job.find('span', class_ = 'srp-skills').text.strip( '\n \r').replace(' ', '')
print(skills)
```

```python
 more_info = job.header.h2.a['href']
more_info
```

```python
published_date = job.find('span', class_ = 'sim-posted').span.text
print(published_date)
```

```python
print(f'''
Company Name: {company_name}
Required Skils: {skills}
Published date: {published_date}
''')
```

```python
for job in jobs:
    published_date = job.find('span', class_ = 'sim-posted').span.text
    if 'few' in published_date:
        company_name = job.find('h3', class_ = 'joblist-comp-name').text.strip()
        skills = job.find('span', class_ = 'srp-skills').text.strip( '\n \r').replace(' ', '')
        more_info = job.header.h2.a['href']
             
        print(f'''
    Company Name: {company_name}
    Required Skils: {skills}
    Published date: {published_date}
    More Info: {more_info}
    ''')
        print()
```
